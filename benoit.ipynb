```python
from Bio.Data.CodonTable import unambiguous_dna_by_id
from Bio.Seq import Seq


def altcodons(codon, table):
    """List codons that code for the same aminonacid / are also stop.

    @param codon
    @table code table id
    @return list of codons

    """
    tab = unambiguous_dna_by_id[table]

    if codon in tab.stop_codons:
        return tab.stop_codons

    try:
        aa = tab.forward_table[codon]
    except:
        return []

    return [
        k
        for (k, v) in tab.forward_table.items()
        if v == aa and k[0] == codon[0] and k[1] == codon[1]
    ]
```

La sequencia utilizada para el desarrollo de las pregunta fue la siguiente:

```python
seq = Seq('ACGCATGCATCA')
```

## a)

```python
def randomCodons(seq, table=1):
    ran = ''
    i = 0
    while i < len(seq):
        cod = seq[i:i+3]
        opciones = altcodons(cod, table)
        newCodon = opciones[random.randint(0, len(opciones)-1)]
        
        ran += newCodon
        i += 3
    return Seq(ran)
```

```python
randomCodons(seq)
```

## b)

```python
from Bio.SeqUtils import GC

def maxCG(seq, table=1):
    maxseq = ''
    i = 0
    while i < len(seq):
        seq_aux = ''
        max_aux = 0
        
        cod = seq[i:i+3]
        opciones = altcodons(cod, table)
        
        for new_seq in opciones:
            if GC(new_seq)>max_aux:
                seq_aux = new_seq
                max_aux = GC(new_seq)
        maxseq += seq_aux
        i += 3
            
    return maxseq
```
