```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
import bioinfo
from copy import copy
```

```python
from Bio.Data.CodonTable import standard_dna_table, unambiguous_rna_by_id, unambiguous_dna_by_id
```

```python
def altcodons(codon, table):
    """List codons that code for the same aminonacid / are also stop.

    @param codon
    @table code table id
    @return list of codons

    """
    tab = unambiguous_dna_by_id[table]

    if codon in tab.stop_codons:
        return tab.stop_codons

    try:
        aa = tab.forward_table[codon]
    except:
        return []

    return [
        k
        for (k, v) in tab.forward_table.items()
        if v == aa and k[0] == codon[0] and k[1] == codon[1]
    ]


def degeneration(codon, table):
    """Determine how many codons code for the same amino acid / are also stop

    @param codon the codon
    @param table code table id
    @param the number of codons also coding for the amino acid codon codes for

    """
    return len(altcodons(codon, table))


def isXdegenerated(x, codon, table):
    """Determine if codon is x-fold degenerated.

    @param codon the codon
    @param table code table id
    @param true if x <= the degeneration of the codon

    """
    return x <= len(altcodons(codon, table))


def degenerated_subseq(seq, x, table):
    """Get a subsequence consisting of the x-fold degenerated codons only."""
    data = ""
    for i in range(0, len(seq), 3):
        codon = seq[i : i + 3]#.tostring()
        if isXdegenerated(x, codon, table):
            data += codon
    return data
```

## a)

```python
def random_codon_replacement(seq, table = 1):
    """ Replaces each codon in seq with a codon that codifies the same aminoacid than the original with equal probability
    among the codos that codify the original"""
    new_seq = copy(seq)
    # Leer la secuencia en saltos de a 3 caracteres
    for i in range(0, len(seq), 3):
        codon = seq[i : i + 3]
        choices = altcodons(codon, table) # obtener los codones que codifican el codon entregado
        if len(choices) != 0: # SI es codificado por algo reemplazar
            replacement = np.random.choice(choices, 1)[0]
            new_seq = new_seq[:i] + replacement + new_seq[i + 3:]
    return new_seq

    
```

## b)
Para tener la mayor cantidad de G y C basta contar la cantidad de C y G en cada una de las opciones posibles para un codon y luego ordenar en orden decreciente usando como llave esa cantidad de de C/Gs:  

```python
def contar_CG(codon):
    dict_freqs = Counter(codon)
    n = 0
    if 'G' in list(dict_freqs.keys()):
        n+=dict_freqs['G']
    if 'C' in list(dict_freqs.keys()):
        n+=dict_freqs['C']
    return n
        
    
def replace_maximizing_CG(seq, table):
    """ Replaces each codon in seq with a codon that codifies the same aminoacid than the original
    maximizing the amount of C & G in the replacements"""
    new_seq = copy(seq)
    # Leer la secuencia en saltos de a 3 caracteres
    for i in range(0, len(seq), 3):
        codon = seq[i : i + 3]
        choices = altcodons(codon, table) # obtener los codones que codifican el codon entregado
        if len(choices) != 0: # SI es codificado por algo reemplazar
            replacement = sorted({cod:contar_CG(cod) for cod in choices}.items(), key = lambda x: x[1], reverse = True)[0][0]
            new_seq = new_seq[:i] + replacement + new_seq[i + 3:]
    return new_seq
    
```

Lo bueno de hacerlo de esta forma es que funciona en ambas direcciones, si se da un porcentaje de GC menor al que ya esta presente en la secuencia el modelo va a intercambiar codones de forma que se disminuya el % a lo más cercano al pedido.

```python
def top_bottom_GC_ORF(record, min_pro_len = 100, table = 1):
    seqs = []
    for strand, nuc in [(+1, record.seq), (-1, record.seq.reverse_complement())]:
        for frame in range(3):
            length = 3 * ((len(record)-frame) // 3) #Multiple of three
            for pro in nuc[frame:frame+length].translate(table).split("*"):
                if len(pro) >= min_pro_len:
                    seqs.append({'pro':pro, 'strand':strand, 'frame': frame, 'len': len(pro), 'porc_GC': contar_CG(pro)/len(pro)*100})
                    
    sorted_seqs_by_GC = sorted(seqs, key = lambda x: x['porc_GC'], reverse = True)
    top_GC = sorted_seqs_by_GC[:3]
    bottom_GC = sorted_seqs_by_GC[-3:]
    print('Top 3 ORFs')
    for seq in top_GC:
        print(f'{seq["pro"][:30]}...{seq["pro"][-3:]} - length {seq["len"]}, strand {seq["strand"]}, frame {seq["frame"]},\
        percentage of GC {round(seq["porc_GC"],2)}%')
        
    print('Bottom 3 ORFs')
    for seq in bottom_GC:
        print(f'{seq["pro"][:30]}...{seq["pro"][-3:]} - length {seq["len"]}, strand {seq["strand"]}, frame {seq["frame"]},\
        percentage of GC {round(seq["porc_GC"],2)}%')
```

```python
def shuffle_nucleotide(file_path, out_filename):
    record = SeqIO.read(file_path, "fasta")
    tmp_seq = list(record.seq)
    np.random.shuffle(tmp_seq)
    record.seq = Bio.Seq.Seq(''.join(tmp_seq))
    SeqIO.write(record, out_filename, "fasta")
```
